# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
#TO DO: Create a function that does a derivative for any function
import numpy as np
import pandas as pd
import sympy as sp


# the derivative function
def dy_dx(expression, variable='x'):
    x = sp.symbols(variable)
    f = sp.sympify(expression)
    derivative = sp.diff(f, x)
    return derivative
````

# Task 2: Test the derivative function written with an equation of your choice
*NB: Must have atleast 2 minimas and atleast 2 maximas*

````python
# let's define x
x = sp.symbols('x')
def f(x):
    return x


# let's define f(x) = x^2 + 5x + 6

def g(x):
    g_x = -x**5 + 3*x**3 - x
    return g_x


# find the derivative of g(x)
der = dy_dx(g(x))
print("The original function is: ", g(f(x)))
print("The derivative is: ", der)
````

# Task 4: Create separate array of Minimas and maximas

````python
def get_max_min(equation):
    x = sp.symbols('x')
    derivative = dy_dx(equation, 'x')
    critical_pts = sp.solve(derivative, x)
    x_critical_pts = []
    y_critical_pts = []

    for point in critical_pts:
        values = equation.subs(x, point)
        x_critical_pts.append(point)
        y_critical_pts.append(values)

    return x_critical_pts, y_critical_pts

y = '-x**5 + 3*x**3 - x'
equation = sp.sympify(y)
arr_minimas, arr_maximas = get_max_min(equation)

print(f'the array of minimas is {arr_minimas} and the array of maximas is {arr_maximas}')
````

````python
x = np.linspace(-2, 2, 100)
y = g(x)

plt.plot(x, y, label='Graph of g(x)')
plt.scatter(arr_minimas, [g(point) for point in arr_minimas], color='red', marker='o', label='Global Minima')
plt.scatter(arr_maximas, [g(point) for point in arr_maximas], color='green', marker='o', label='Global Maxima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Equation with Global Minima and Global Maxima illustrated')
````
